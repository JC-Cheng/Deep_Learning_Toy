```python
from __future__ import print_function
from scipy.misc import imresize, imread

import numpy as np
import theano
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
w = np.random.randn(3, 2)
W = theano.shared(np.zeros((3, 2)))

print(W.get_value())

W.set_value(w)
print(W.get_value())

# You can load pre-trained models and assign parameters by doing this
```

```python
# the classifier will predict top 5 ImageNet Class
# see http://image-net.org/

from sklearn_theano.feature_extraction import OverfeatClassifier
clf = OverfeatClassifier(top_n=5)
```

```python
def load_picture_and_classify(path, pixel=[231, 231]):
    img = imresize(imread(path), pixel)
    plt.imshow(img)
    return clf.predict(img)
```
